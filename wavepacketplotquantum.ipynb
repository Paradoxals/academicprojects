```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
alpha = 2.0
k0 = 10.0
N = np.sqrt(alpha)  # From normalization

# Create x and k ranges
x = np.linspace(-5, 5, 1000)
k = np.linspace(5, 15, 1000)

# 1. Calculate Re[(x)]
# (x) = N*e^(-|x|/2)*e^(ikx)
# Re[(x)] = N*e^(-|x|/2)*cos(kx)
psi_real = N * np.exp(-alpha * np.abs(x) / 2) * np.cos(k0 * x)

# 2. Calculate |(x)|
# |(x)| = N*e^(-|x|)
psi_squared = N**2 * np.exp(-alpha * np.abs(x))

# 3. Calculate |A(k)|
# A(k) = ((/2))/(2(/4 + (k-k)))
A_k = (alpha * np.sqrt(alpha/2)) / (2 * np.pi * (alpha**2/4 + (k0 - k)**2))
A_k_squared = A_k**2

# Create the plots
fig, axes = plt.subplots(3, 1, figsize=(10, 10))

# Plot 1: Re[(x)]
axes[0].plot(x, psi_real, 'b-', linewidth=2)
axes[0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[0].grid(True, alpha=0.3)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('Re[(x)]', fontsize=12)
axes[0].set_title('Real Part of Wave Function: Re[(x)] = Ne^(-|x|/2)cos(kx)', fontsize=13, fontweight='bold')

# Plot 2: |(x)|
axes[1].plot(x, psi_squared, 'r-', linewidth=2)
axes[1].fill_between(x, 0, psi_squared, alpha=0.3, color='r')
axes[1].grid(True, alpha=0.3)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('|(x)|', fontsize=12)
axes[1].set_title('Position Space Probability Density: |(x)| = Ne^(-|x|)', fontsize=13, fontweight='bold')

# Plot 3: |A(k)|
axes[2].plot(k, A_k_squared, 'g-', linewidth=2)
axes[2].fill_between(k, 0, A_k_squared, alpha=0.3, color='g')
axes[2].axvline(x=k0, color='k', linestyle='--', alpha=0.5, label=f'k = {k0}')
axes[2].grid(True, alpha=0.3)
axes[2].set_xlabel('k', fontsize=12)
axes[2].set_ylabel('|A(k)|', fontsize=12)
axes[2].set_title('Momentum Space Probability Density: |A(k)|', fontsize=13, fontweight='bold')
axes[2].legend()

plt.tight_layout()
plt.savefig('wave_packet_plots.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Parameters used:")
print(f" = {alpha}")
print(f"k = {k0}")
print(f"N =  = {N:.3f}")
```
